```python
import pandas as pd
movies = pd.read_csv("fandango_score_comparison.csv")
movies
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.hist(movies["Metacritic_norm_round"],10,(0,5.0))
plt.show()
plt.hist(movies["Fandango_Stars"],10,(0,5.0))
plt.show()
```

### Comparing Metacritic and Fandango movie scores

The main difference is that Fandango has virtually no movies rated below a 3 out of 5 with the majority of movies falling towards the high end of the scale. Metacritic has a much broader range of ratings distributed much more equally from about 1.5 to 5.

```python
import numpy

f_mean = movies["Fandango_Stars"].mean()
m_mean = movies["Metacritic_norm_round"].mean()
f_median = movies["Fandango_Stars"].median()
m_median = movies["Metacritic_norm_round"].median()
f_std = movies["Fandango_Stars"].std()
m_std = movies["Metacritic_norm_round"].std()

print("Mean:   ",f_mean,m_mean)
print("Median: ",f_median,m_median)
print("Std Dev:",f_std,m_std)
```

### Comparing Fandango and Metacritic Methodology
Fandango is unclear on their rating system and might not even keep their own anymore. Metacritic is clear that it uses a normalized weighted average of critic's reviews to get their scores. 

### Comparing Fandango and Metacritic Stats
Fandango has much higher mean and median and a lower standard deviation indicating Fandango's scores are clustered more tightly around higher values.

Metacritic has a broader range of scores (due to larger std dev) with lower average values.

Fandago's scores may be inflated because they also are looking to sell consumers movie tickets. Their scoring process is also less transparent.

```python
plt.scatter(movies["Fandango_Stars"],movies["Metacritic_norm_round"])
```

```python
movies["fm_diff"] = (movies["Fandango_Stars"] - movies["Metacritic_norm_round"]).abs()
movies = movies.sort_values("fm_diff", ascending = False)
movies.head(5)
```

```python
from scipy.stats import pearsonr

r, p_value = pearsonr(movies["Fandango_Stars"],movies["Metacritic_norm_round"])
r
```

### Correlation between Fandango and Metacritic scores

The r correlation coefficient between these scores is rather low. This means that a rating on one site is not a good indicator of what the rating for a movie on the other site will be. Not only are Fandango's scores inflated, but they also don't seem to inflate all movies equally.

```python
from scipy.stats import linregress
m, b, r, p, stderr = linregress(movies["Metacritic_norm_round"],movies["Fandango_Stars"])
pred_3 = 3 * m + b
pred_3
```

## Finding Residuals

```python
pred_1 = 1*m+b
pred_5 = 5*m+b
print(pred_1,pred_5)
```

```python
plt.scatter(movies["Metacritic_norm_round"],movies["Fandango_Stars"])
plt.plot([1,5],[pred_1,pred_5])
plt.xlim(1,5)
plt.show()
```
